# Problema de ingeniería

**Calcular el área de un terreno rectangular.**

Este cálculo es muy común cuando se necesita estimar:

* área de cultivo
* área de una losa
* superficie de un lote

La fórmula es:

[
Area = Largo \times Ancho
]

---

# 1. Definir Entradas, Proceso y Salidas

### Entradas (Inputs)

Datos que el usuario debe proporcionar.

* `largo` del terreno (m)
* `ancho` del terreno (m)

### Proceso

Aplicar la fórmula matemática:

[
area = largo * ancho
]

### Salida (Output)

* `area` del terreno en **metros cuadrados (m²)**

---

# 2. Algoritmo (explicación paso a paso)

1. Pedir al usuario el **largo del terreno**.
2. Pedir al usuario el **ancho del terreno**.
3. Multiplicar largo por ancho.
4. Mostrar el resultado del área.

---

# 3. Pseudocódigo

```
INICIO

Leer largo
Leer ancho

area = largo * ancho

Mostrar area

FIN
```

---

# 4. Código en Python

```python
# Programa para calcular el área de un terreno rectangular

# Entrada de datos
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

### Un primer acercamiento al código

Imagina que necesitas calcular el **volumen de agua necesario para un tanque cilíndrico** en un proyecto de riego. Antes de escribir el código, debemos pensar en la lógica:

1. Definir las dimensiones.
2. Aplicar la fórmula geométrica.
3. Mostrar el resultado.

Aquí tienes un ejemplo siguiendo nuestras **buenas prácticas** (variables descriptivas y modularización):

````python
import math

def calcular_volumen_tanque(radio_m, altura_m):
    """
    Calcula el volumen de un cilindro en metros cúbicos.
    """
    # Usamos la fórmula: Área de la base * Altura
    volumen_m3 = math.pi * (radio_m**2) * altura_m
    return volumen_m3

# Datos de entrada (Imagina que vienen de un sensor o un formulario)
radio_tanque = 2.5  # metros
altura_tanque = 5.0 # metros

# Llamada a la función
resultado = calcular_volumen_tanque(radio_tanque, altura_tanque)

print(f"El volumen del tanque es: {resultado:.2f} m³")
````
